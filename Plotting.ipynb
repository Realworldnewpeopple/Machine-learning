```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df=pd.read_excel("data.xlsx",convert_float=False)
```

```python
df.index=df['Unit']
```

```python
df.info()
```

```python
df.head()
```

```python
df=df.drop(df.columns[0],axis=1)
```

```python
df.head()
```

```python
df1=df
df1['Employees working very long hoursInformation on item']*=1000
df1['Life satisfactionInformation on item']*=1000
```

```python
df1['Household net adjusted disposable incomeInformation on item'].plot(kind='line',figsize=(20,10),label='Income')
df1['Employees working very long hoursInformation on item'].plot(kind='line',figsize=(20,10),label='Working Hours * 1000')
df1['Life satisfactionInformation on item'].plot(kind='line',figsize=(20,10),label='Life Satisfaction * 1000')

new_labels=['Australia', 'Austria', 'Belgium', 'Canada', 'Chile', 'Czech Republic',
       'Denmark', 'Estonia', 'Finland', 'France', 'Germany', 'Greece',
       'Hungary', 'Iceland', 'Ireland', 'Israel', 'Italy', 'Japan', 'Korea',
       'Latvia', 'Luxembourg', 'Mexico', 'Netherlands', 'New Zealand',
       'Norway', 'Poland', 'Portugal', 'Slovak Republic', 'Slovenia', 'Spain',
       'Sweden', 'Switzerland', 'Turkey', 'United Kingdom', 'United States',
       'OECD - Total', '   Brazil', '   Russia', '   South Africa']
plt.xticks(ticks=np.arange(len(df.index)),labels=new_labels,rotation=90, fontsize=9)
plt.legend(fontsize=20)
plt.show()
```

```python
df.hist(bins=50,figsize=(40,40))
plt.show()
```

```python
plt.figure(figsize=(20,10))
plt.scatter(df['Life satisfactionInformation on item'],df['Household net adjusted disposable incomeInformation on item'],marker='o',label="Income vs Life Satisfaction rating")
plt.xlabel("Life Satisfaction Rating",fontsize=25)
plt.ylabel("Household Income",fontsize=25)
z=df['Life satisfactionInformation on item'].values
y=df['Household net adjusted disposable incomeInformation on item'].values
x=df.index.values
for i, txt in enumerate(x):
    plt.annotate(txt,(z[i],y[i]),fontsize=15)
plt.legend(loc='best',fontsize=15)
plt.show()
```
