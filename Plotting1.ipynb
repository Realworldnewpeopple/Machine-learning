```python
x_val=np.c_[df['Household net financial wealthInformation on item']]
y_val=np.c_[df['Life satisfactionInformation on item']]
```

```python
lnr=LinearRegression()
```

```python
lnr.fit(x_val,y_val)
```

```python
lnr.score(x_val,y_val)
```
